# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 50
sample_std = 5
n = 100 # assuming a sample size of 100

# calculate the standard error of the mean
sem = sample_std / (n ** 0.5)

# calculate the margin of error
moe = stats.t.ppf(0.975, n-1) * sem

# calculate the confidence interval
ci = (sample_mean - moe, sample_mean + moe)

# print the results
print("The 95% confidence interval is", ci)
```

# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

observed_frequencies = np.array([18, 24, 17, 10, 11, 20])  # observed frequencies
expected_frequencies = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * sum(observed_frequencies)  # expected frequencies

chi_sq, p = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

if p < 0.05:
    print(f"The chi-square test statistic is {chi_sq:.3f} and the p-value is {p:.3f}.")
    print("The distribution of M&M colors in the bag does not match the expected distribution.")
else:
    print(f"The chi-square test statistic is {chi_sq:.3f} and the p-value is {p:.3f}.")
    print("The distribution of M&M colors in the bag matches the expected distribution.")
```

# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 100

alpha = 0.1 # 90% confidence level
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)

margin_of_error = t_critical * sample_std / (sample_size ** 0.5)

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("90% confidence interval:", confidence_interval)
```

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom
df = 10

# Create a range of chi-square values
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, chi2.pdf(x, df))
ax.set_xlabel('Chi-square value')
ax.set_ylabel('Probability density')
ax.set_title('Chi-square distribution with 10 degrees of freedom')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(chi2.ppf(0.01, df), 15, 100)
ax.fill_between(x_fill, chi2.pdf(x_fill, df), alpha=0.5)
ax.annotate('Chi-square statistic = 15', xy=(15, 0.02), xytext=(17, 0.1),
            arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```

To calculate the 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the following formula:

CI = p  zsqrt((p(1-p))/n)

where:

p = proportion of people in the sample who prefer Coke
z = z-score corresponding to the confidence level (in this case, 2.576 for a 99% confidence level)
n = sample size

First, let's calculate the proportion of people in the sample who prefer Coke:

p = 520/1000
p = 0.52

Next, we can calculate the margin of error:

ME = zsqrt((p(1-p))/n)
ME = 2.576sqrt((0.52(1-0.52))/1000)
ME = 0.049

Finally, we can calculate the confidence interval:

CI = 0.52  0.049
CI = (0.471, 0.569)

Interpretation: We can say with 99% confidence that the true proportion of people in the population who prefer Coke is between 0.471 and 0.569.







# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# state null and alternative hypotheses
# null hypothesis: observed frequencies follow expected frequencies of a fair coin
# alternative hypothesis: observed frequencies do not follow expected frequencies of a fair coin
null_hypothesis = "observed frequencies follow expected frequencies of a fair coin"
alternative_hypothesis = "observed frequencies do not follow expected frequencies of a fair coin"

# level of significance (alpha)
alpha = 0.05

# observed frequencies
observed_frequencies = np.array([55, 45])

# expected frequencies
expected_frequencies = np.array([50, 50])

# calculate chi-square statistic and p-value
chi_square, p_value = chisquare(observed_frequencies, expected_frequencies)

# calculate critical value
critical_value = 3.84

# compare chi-square statistic to critical value and print result
if chi_square > critical_value:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
    
# calculate 95% confidence interval for the true proportion of tails
n = 100
p_hat = 0.45
std_error = np.sqrt(p_hat * (1 - p_hat) / n)
z_alpha_half = 2.576
lower_bound = p_hat - z_alpha_half * std_error
upper_bound = p_hat + z_alpha_half * std_error
print("95% confidence interval: [{:.4f}, {:.4f}]".format(lower_bound, upper_bound))
```
