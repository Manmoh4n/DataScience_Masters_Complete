# Matplotlib

Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.plot(x,y)
plt.show()
```

```python
#Scatter plot
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
plt.title("This is title")
plt.xlabel("This is x label")
plt.ylabel("This is y label|")
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes =700* np.random.rand(50)
plt.scatter(x,y, c=colours, s=sizes , alpha=.8) #alpha represent the intensity of color
plt.title("This is title")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
```

```python
x= np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.figure(figsize=(6,2))  #It is use to control the fig size 
plt.plot(x,y,'--r')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colours = np.random.rand(50)
sizes =700* np.random.rand(50)
plt.scatter(x,y, c='b', s=sizes , alpha=.8, marker="v") #alpha represent the intensity of color
plt.title("This is title")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
```

```python
# Data represent in 3D
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()

```
